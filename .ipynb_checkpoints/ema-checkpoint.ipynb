```python
from matplotlib.patches import Ellipse

def draw_ellipse(position, covariance, ax=None, color='red', **kwargs):
    """Draw an ellipse with a given position and covariance"""
    ax = ax or plt.gca()
    
    # Convert covariance to principal axes
    if covariance.shape == (2, 2):
        U, s, Vt = np.linalg.svd(covariance)
        angle = np.degrees(np.arctan2(U[1, 0], U[0, 0]))
        width, height = 2 * np.sqrt(s)
    else:
        angle = 0
        width, height = 2 * np.sqrt(covariance)
    
    # Draw the Ellipse
    for nsig in range(1, 2):
        ax.add_patch(Ellipse(position, nsig * width, nsig * height,
                             angle, color=color, **kwargs))
        
def plot_gmm(gmm, X, label=True, ax=None, color='red'):
    ax = ax or plt.gca()
    labels = gmm.predict(X)
    if label:
        ax.scatter(X[:, 0], X[:, 1], s=2, zorder=2, alpha=0.05, marker='x', color=color)
    else:
        pass
    ax.axis('equal')
    
    w_factor = 0.5 / gmm.weights_.max()
    for pos, covar, w in zip(gmm.means_, gmm.covariances_, gmm.weights_):
        if w == gmm.weights_.max():
          draw_ellipse(pos, covar, color='green', alpha=w * w_factor)
        else:
          draw_ellipse(pos, covar, color=color, alpha=w * w_factor)
```

```python
fig, ax = plt.subplots(1, figsize=(15,15))

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title("Crime data with GMM ellipses")

plot_gmm(gmm_street, X_street, label=True, color='blue')
plot_gmm(gmm_ss, X_ss, label=True, color='red')
ax.set_xlim([-2.2, -1.2])
ax.set_ylim([53.4, 54.2])
# plt.scatter(X_ss[:,0], X_ss[:,1], marker='+')
```

```python
def gmm_kl(gmm_p, gmm_q, n_samples=10**5):
    X, _ = gmm_p.sample(n_samples)
    log_p_X = gmm_p.score_samples(X)
    log_q_X = gmm_q.score_samples(X)
    return log_p_X.mean() - log_q_X.mean()
```

#KL Divergence comparison

```python
df_random_street = pd.DataFrame([])
df_random_ss = pd.DataFrame([])
with zipfile.ZipFile(THE_FILE) as z:
  for i in tqdm(z.namelist()):
    if 'cambridgeshire' in i:
      if i.endswith('street.csv'):
        df_random_street = df_random_street.append(pd.read_csv(z.open(i)))
      if i.endswith('search.csv'):
        df_random_ss = df_random_ss.append(pd.read_csv(z.open(i)))
        
nans = df_random_street[['Latitude', 'Longitude']].notnull().values
nans = nans[:,0] * nans[:,1]
df_random_street = df_random_street[nans]
X_random_street = df_random_street[['Longitude', 'Latitude']].values

nans = df_random_ss[['Latitude', 'Longitude']].notnull().values
nans = nans[:,0] * nans[:,1]
df_random_ss = df_random_ss[nans]
X_random_ss = df_random_ss[['Longitude', 'Latitude']].values

first_mask = (X_random_ss[:,1] > np.min(X_random_street[:,1])) * (X_random_ss[:,1] < np.max(X_random_street[:,1]))
df_random_ss = df_random_ss[first_mask]
X_random_ss = X_random_ss[first_mask]

second_mask = (X_random_ss[:,0] > np.min(X_random_street[:,0])) * (X_random_ss[:,0] < np.max(X_random_street[:,0]))
df_random_ss = df_random_ss[second_mask]
X_random_ss = X_random_ss[second_mask] 

gmm_random_street = GaussianMixture(n_components=N)
gmm_random_ss = GaussianMixture(n_components=N)
gmm_random_street.fit(X_random_street)
gmm_random_ss.fit(X_random_ss)

print(gmm_kl(gmm_random_ss, gmm_random_street, n_samples=10**5))
```

```python
fig, ax = plt.subplots(1, figsize=(15,15))

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title("Gaussian ellipses of  ")

plot_gmm(gmm_random_street, X_random_street, label=True, color='blue')
plot_gmm(gmm_random_ss, X_random_ss, label=True, color='red')
ax.set_xlim([-0.5, 0.5])
ax.set_ylim([52, 52.8])
```

```python
fig, ax = plt.subplots(1, figsize=(15,15))
plot_gmm(gmm_random_street, X_random_street, label=False, color='blue')
plot_gmm(gmm_random_ss, X_random_ss, label=False, color='red')
ax.set_xlim([-0.5, 0.5])
ax.set_ylim([52, 52.8])
plt.show()
```
